```python
ratings_with_books=rating.merge(ds,on='ISBN')
```

```python
ratings_with_books.shape
```

```python
num_rating=ratings_with_books.groupby('title')['Brating'].count().reset_index()
```

```python
num_rating
```

```python
num_rating.rename(columns={'Brating':'number of ratings'},inplace=True)
```

```python
finalrating=ratings_with_books.merge(num_rating,on='title')
```

```python
finalrating
```

```python
finalrating.shape
```

```python
finalrating=finalrating[finalrating['number of ratings']>=50]
```

```python
finalrating.shape
```

```python
finalrating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
finalrating.shape
```

```python
book_pivot=finalrating.pivot_table(columns='user_id',index='title',values='Brating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors as nn
```

```python
model=nn(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions=model.kneighbors(book_pivot.iloc[241,:].values.reshape(1,-1))
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def recommendation(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances, suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1))
    
    for i in range(len(suggestions)):
        if i==0:
            print("The suggestions for",book_name,"are : ")
        if not i:
            print(book_pivot.index[suggestions[i]])
    
```

```python
recommendation("Harry Potter and the Sorcerer's Stone (Book 1)")
```
